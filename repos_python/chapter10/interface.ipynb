```python
# 버튼으로 레이블 값을 변경하는 이벤트 처리
from tkinter import *

# 현재 값을 읽으려면 해당 변수에 대한 get() 메서드를 호출
def add():
    n = int(count.get())
    n += 1
    count.set(n)

# 값을 변경하려면 set() 메서드를 호출
def sub():
    n = int(count.get())
    n -= 1
    count.set(n)


str = ['add', 'substraction']
win = Tk()

count = StringVar(value='0')

data = Label(win, width=20, textvariable=count)
data.grid(row=0, column=0, columnspan=2)

# *주의사항 함수명 이후에 () 없음
badd = Button(win, text=str[0], command=add)
badd.grid(row=1, column=0)
bsub = Button(win, text=str[1], command=sub)
bsub.grid(row=1, column=1)

win.mainloop()
```

```python
# 버튼 이벤트 처리로 섭씨 온도를 화씨 온도로 전환
from tkinter import *

def fahrenheit(celsius):
    return celsius * (9 / 5) + 32

def clicked():
    # eval('expression')은 실행 가능한 연산식 문자열인 expression을 실행한 결과를 반환함.
    degree = eval(cel.get().strip())
    print(degree)
    fah.delete(0, END)
    fah.insert(0, str(fahrenheit(degree)))

win = Tk()
win.title('conversion celsius into fahrenheit')

lbcel = Label(win, text='celsius temperature', width=20)
lbfar = Label(win, text='fahrenheit temperature')
lbcel.grid(row=0, column=0)
lbfar.grid(row=1, column=0)

cel = Entry(win, width=20)
fah = Entry(win)
cel.grid(row=0, column=1)
fah.grid(row=1, column=1)

cvt = Button(win, text='conversion', width=20, command=clicked)
cvt.grid(row=2, column=0, columnspan=2)

win.mainloop()
```

```python
# 위젯의 bind() 멤소드로 마우스 이벤트에 직선 그리기
from tkinter import *

win = Tk()
win.title('Draw a line on the canvas')
win.geometry('300x300')

def click(event):
    global sX, sY
    print(f'click location', event.x, event.y)
    sX, sY = event.x, event.y

def release(event):
    global eX, eY
    print(f'release location', event.x, event.y)
    eX, eY = event.x, event.y
    canvas.create_line(sX, sY, eX, eY, fill='blue', width=2)

canvas = Canvas(win, relief='solid', bd=2)
canvas.pack(expand=True, fill='both')

canvas.bind('<Button-1>', click)
canvas.bind('<ButtonRelease-1>', release)

win.mainloop()
```

```python
import pygame

# 색상 정의
WHITE = (255, 255, 255)
# 윈도 크기 정의
size = (300, 200)

# 윈도 초기화
pygame.init()

# 화면 크기 지정해 스크린을 생성
screen = pygame.display.set_mode(size)
# 제목인 캡션 지정
pygame.display.set_caption('first pygame win!')

# 윈도 종료 플래그로 사용되는 변수의 초기값 지정
done = False
# 메인 루프
while not done:
    # 입력받은 키(이벤트)를 가져옴
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            done = True

    screen.fill(WHITE)
    pygame.display.update()
# 종료 버튼
pygame.quit()
```

```python
# animation
import pygame as p, random as r

# 게임 엔진 초기화
p.init()

BLACK = [0, 0, 0]
WHITE = [255, 255, 255]
SIZE = [300, 300]
SNOW_CNT = 70

screen = p.display.set_mode(SIZE)
p.display.set_caption('a snowy night')

# a snowy night
snow_list = []

# snow_list에 SNOW_CNT수만큼의 좌표 저장
for i in range(SNOW_CNT):
    x = r.randrange(0, SIZE[0])
    y = r.randrange(0, SIZE[1])
    snow_list.append([x,y])
# FPS 프레임 속도
clock = p.time.Clock()

# 메인 루프
done = False
while not done:
    for event in p.event.get():
        if event.type == p.QUIT:
            done = True

    screen.fill(BLACK)

    # 눈 내리는 모습 그리기
    for i in range(len(snow_list)):
        # 눈 모양의 원 그리기
        radius = 1
        # radius = r.randint(1, 3)
        p.draw.circle(screen, WHITE, snow_list[i], radius)

        # 눈 snow_list[i]의 y좌표를 1증가시켜 눈이 내리도록 수정
        snow_list[i][1] += 1
        # snow_list[i][1] += r.randint(1, 3)
        # 눈 snow_list[i]의 x좌표를 (-1, 0, 1) 중 하나에서 선택, 이전 값에 더해 대입
        snow_list[i][0] += r.randint(-1, 1)

        # snow_list[i]가 윈도 바닥으로 내려가면
        if snow_list[i][1] > SIZE[1]: #snow_list[i][1]은 snow_list[i] 눈의 y좌표
            # y좌표를 위(음수)로 다시 지정
            snow_list[i][1] = r.randrange(-5, 0)
            # x좌표를 다시 지정, snow_list[i][0]은 snow_list[i] 눈의 x좌표
            snow_list[i][0] = r.randrange(0, SIZE[0])
    p.display.update()
    # 초당 수정될 프레임 수 지정, 초당 20 프레임 화면이 수정됨
    clock.tick(20)
p.quit()
```

```python
# 5. pygame 윈도 바탕 색상이 1초에 한번씩 변하면서 색상 이름 출력
#    윈도 색상과 폰트의 색상이 같으면 표시 글자가 보이지 않으므로 폰트의 색상은
#    위 순서의 이웃한 다음색으로 지정해 출력
import pygame as p
import time

size = (190, 100)

p.init()
screen = p.display.set_mode(size)   #스크린 생성
p.display.set_caption('pygame')     #캡션 지정

rgbs = [(0, 0, 0), (255, 0, 0), (0, 255, 0), (0, 0, 255), (255, 255, 0)]
colors = ['black', 'red', 'green', 'blue', 'yellow']

index = 0
done = False
# 메인 루프
while not done:
    for event in p.event.get():
        if event.type == p.QUIT:
            done = True
    
    # 바탕 색상
    screen.fill(rgbs[index])
    time.sleep(1)
    
    # 폰트 색상
    next_color = rgbs[(index + 1) % len(rgbs)]
    font = p.font.Font('freesansbold.ttf', 20)
    text = colors[index]
    text_surface = font.render(text, True, next_color)
    text_rect = text_surface.get_rect()
    text_rect.center = (size[0] // 2, size[1] // 2)
    screen.blit(text_surface, text_rect)

    p.display.update()

    index = (index + 1) % len(rgbs)
p.quit()
```

```python
# 6. 다음을 참고해 마우스를 누른채 이동하면 붉은색의 작은 원이 계속 그려지는 pygame 작성
#    마우스를 누른채 이동하는 위치를 리스트 mpos에 저장
#    변수 mdown은 마우스를 누른채 이동하는 동안 True인 변수
import pygame as p

size = (640, 480)

p.init()
screen = p.display.set_mode(size)
p.display.set_caption('pygame')

mpos = []
mdown = False

# 메인 루프
while True:
    for event in p.event.get():
        if event.type == p.QUIT:
            exit()
        elif event.type == p.MOUSEBUTTONDOWN:
            mdown = True
        elif event.type == p.MOUSEBUTTONUP:
            mdown = False
        elif event.type == p.MOUSEMOTION and mdown:
            mpos.append(event.pos)

    # 화면을 흰색으로 지운다
    screen.fill((255, 255, 255))

    # 원을 그린다
    for pos in mpos:
        p.draw.circle(screen, (255, 0, 0), pos, 2)

    # 화면을 업데이트한다
    p.display.update()
```
