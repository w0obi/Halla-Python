```python
import math as m, random as rd

def getarea():
    return m.pi * (r**2)

r = (round(rd.random()* 10, 2))
print(f'원의 반지름: {r}')
print(f'원주율 pi: {m.pi:.4f}')
print(f'반지름 {r}인 원의 면적은 {getarea():.2f}')
```

```python
import math as m, statistics as s
enter = (1, 6, 11, 16)

def ftl():
    for i in enter:
        ft = m.factorial(i)
        print(f'{i}! = {ft}')

ftl()
print()

st = [80, 99, 77, 65, 92, 74, 82]
if __name__ == '__main__':
    print(st)
    print(f'중앙 값: {s.median(st):.2f}')
    print(f'평균: {s.mean(st):.2f}')
    print(f'분산: {s.variance(st):.2f}')
    print(f'표준편차: {s.stdev(st):.2f}')
```

```python
import turtle as t

def drawpolygon(n, size):
    for i in range(n):
        t.forward(size)
        t.right(360/n)

t.setup(500, 500)
t.speed(1)
t.pu()
t.goto(-100, 0)
t.pd()
        
drawpolygon(5, 50)

t.exitonclick()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi*2, np.pi*2, 720)
cos, sin = np.cos(x), np.sin(x)

plt.plot(x, sin, ls=':', label='sin')
plt.plot(x, cos, ls='--', label='cos')

plt.title("Graph of sin cos")
plt.xlabel('radians')
plt.ylabel('value')
plt.grid()
plt.legend()
plt.show()
```
