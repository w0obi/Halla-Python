```python
# 거북이 wasd 방향으로 10씩 이동, turtle의 listen()함수를 호출하여 이벤트가 작동하도록 함.
import turtle as t

def goright():
    t.setheading(0)
    t.fd(10)
    
def goleft():
    t.setheading(180)
    t.fd(10)
    
def goup():
    t.setheading(90)
    t.fd(10)                               # 거북이 앞으로 이동
    
def godown():
    t.setheading(270)
    t.fd(10)                               # 거북이 앞으로 이동
    
def cls():
    t.clear()

t.setup(500, 500)                              # 창의 크기 지정
t.shape('turtle')                              # 제목 turtle
t.clear()                                      # 화면 지우기
t.home()                                       # 원점으로 이동

t.onkeypress(goright, 'Right')
t.onkeypress(goleft, 'Left')
t.onkeypress(goup, 'Up')
t.onkeypress(godown, 'Down')
t.onkeypress(cls, 'Escape')
t.listen()
t.exitonclick()
```

```python
import random as r
import turtle as t

cols = ['red', 'blue', 'green', 'purple', 'magenta',
        'gray', 'yellow', 'cyan', 'orange', 'aqua']

width, height = 500, 500                       

def checkround():
    def round(x, y):
        t.ht(); t.pu()                         # ht = hideturtle, pu = pen up
        t.goto(x, y)
        t.pd(); t.st()                         # pd = pen down, st = show turtle
        t.pencolor(cols[r.randint(0, len(cols)-1)])     # pen의 컬러를 랜덤 설정

    x, y = t.pos()                             # 현재 거북이 위치를 반환

    # x좌표 벗어나면 그 반대로 이동
    if not (-width/2 < x < width/2):
        round(-x, y)

    # y좌표를 벗어나면 그 반대로 이동
    if not (-height/2 < y < height/2):
        round(x, -y)

def goright():
    t.setheading(0)
    t.fd(10)
    checkround()

def goleft():
    t.setheading(180)
    t.fd(10)
    checkround()

def goup():
    t.setheading(90)
    t.fd(10)                               # 거북이 앞으로 이동
    checkround()

def godown():
    t.setheading(270)
    t.fd(10)                               # 거북이 앞으로 이동
    checkround()

def cls():
    t.clear()

t.setup(width, height)                              # 창의 크기 지정
t.shape('turtle')                              # 제목 turtle
t.clear()                                      # 화면 지우기
t.home()                                       # 원점으로 이동

t.onkeypress(goright, 'Right')
t.onkeypress(goleft, 'Left')
t.onkeypress(goup, 'Up')
t.onkeypress(godown, 'Down')
t.onkeypress(cls, 'Escape')
t.listen()
t.exitonclick()
```

- 엔트리로 두 수 a, b를 입력받음
    - 레이블로 결과 값 나타냄
    - 나머지는 버튼_아래쪽에 나타내기

```python
from tkinter import *

def calculate(*args):
    try:
        value = float(feet.get())
        meters.set((0.3048 * value * 10000.0 + 0.5)/10000.0)
    except ValueError as e:
        print('feet 입력에 문제가 있습니다.')
        print(f'에러 발생 이름{type(e)}')
        print(f'에러 발생 이유{e}')
        feet.set('다시 입력하세요!')
    else:
        print('feet 입력이 잘 됐습니다')
        print(f'feet 입력 값: {value}')
    finally:
        print(f'입력 예외 처리가 잘 되는군요!')

root = Tk()
root.title('feet to meters')

mainframe = Frame(root, padx=10, pady= 5)
mainframe.grid(column=0, row=0)

feet = StringVar()
meters = StringVar()

entfeet = Entry(mainframe, textvariable=feet, width=14, )
lbMeter = Label(mainframe, textvariable=meters, relief='raised', )
entfeet.grid(column=0, row=0, sticky=(W, E))
lbMeter.grid(column=0, row=1, sticky=(W, E))

Label(mainframe, text='feet').grid(column=1, row=0, sticky=W)
Label(mainframe, text='meters').grid(column=1, row=1, sticky=W)

calcBtn = Button(mainframe, text='Calculate', width=36, command=calculate)
calcBtn.grid(column=0, row=3, columnspan=3, sticky=W)

for child in mainframe.winfo_children():
    child.grid_configure(padx=5, pady=3)

entfeet.focus()
root.bind('<Return>', calculate)

root.mainloop()
```
